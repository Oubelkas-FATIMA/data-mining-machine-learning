## coordinates and diagonals
---

```python
#controlling axis and adding diagonals
g = sns.FacetGrid(movies, row='Genre', col='Year',hue="Genre")
kws = dict(s=50, linewidth = 0.5, edgecolor='black') # dictionary with keywords in order to make the plot more clear
g = g.map(plt.scatter,'RottenTomatoesRatings','AudienceRating',**kws)
g.set(xlim=(0,100), ylim=(0,100)) #setting axis from 0 to 100 each

# now we need to add diagonal to each one of the plots
for ax in g.axes.flat: #check theses functions from shift+tab
    ax.plot((0,100),(0,100), c="gray",ls="--")
# each dote is represnting a movie

g.add_legend()
```

## styling dashoard 
---

```python
sns.set_style('dark', {"axes.facecolor":"black"}) #  because we don't have much choices for colors in seaborn so we have to
#pass a dictionary with the wanted color 
# we want to make all background as black
f, axes = plt.subplots(2,2,figsize=(15,15))  

#fig 1
k1= sns.kdeplot(movies.BudgetMillion,\
                shade=True ,\
                shade_lowest=True,\
                 cmap="inferno",\
                ax=axes[0,0],cbar=True)
#fig 2
k2= sns.kdeplot(movies.RottenTomatoesRatings,\
                shade=True ,\
                shade_lowest=True,\
                 ax=axes[0,1])

#fig 3
z= sns.violinplot(data= movies[movies.Genre=="Drama"], x="Year" , y="RottenTomatoesRatings",ax=axes[1,0],\
                 palette='YlOrRd')


k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
```
